```python

class ODEF(nn.Module):
    def __init__(self, dim, m_dim):
        super().__init__()
        self.A = nn.Linear(dim, dim, bias=False)
        self.B = nn.Linear(dim, dim, bias=True)
        self.W = nn.Linear(dim, dim, bias=False)
        self.m_proj = nn.Linear(m_dim, dim)  # project perturbation emb to gene space

    def forward(self, t, x, m):
        m_proj = self.m_proj(m)                      # shape: (batch, dim)
        activated = F.relu(self.B(x))                # σ(B·x + β)
        stimulation = self.A(activated) * m_proj     # A·σ(...) ⊙ M
        decay = self.W(x)                            # W·x
        return stimulation - decay

class ODEWrapper(nn.Module):
    def __init__(self, odefunc, m):
        super().__init__()
        self.odefunc = odefunc
        self.m = m

    def forward(self, t, x):
        return self.odefunc(t, x, self.m)



class PerturbedNeuralODE(nn.Module):
    def __init__(self, input_dim, output_dim, control_expr_dim, hidden_dim=128):
        super().__init__()
        self.control_proj = nn.Linear(control_expr_dim, input_dim)  # NEW
        self.input_layer = nn.Linear(input_dim, hidden_dim)
        self.odefunc = ODEF(hidden_dim, input_dim)
        self.output_layer = nn.Linear(hidden_dim, output_dim)


    def forward(self, pert_emb, control_expr):
        control_proj = self.control_proj(control_expr)  # [batch, input_dim]
        x0 = self.input_layer(control_proj)
        m = pert_emb
        ode_wrapper = ODEWrapper(self.odefunc, m)
        t = torch.tensor([0., 1.], device=x0.device)
        out = odeint(ode_wrapper, x0, t, method='dopri5')[1]
        return self.output_layer(out)


train_dataset = PerturbationDataset(
    adata=pert_data.adata,
    set2conditions=pert_data.set2conditions,
    split='train',
    gene_emb_matrix=gene_emb_matrix,
    pert_emb_matrix=pert_emb_matrix,
    node_map=pert_data.node_map,
    node_map_pert=pert_data.node_map_pert
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

node_model = PerturbedNeuralODE(
    input_dim=pert_emb_matrix.shape[1],               # perturbation embedding dim
    output_dim=pert_data.adata.shape[1],              # gene count (output size)
    control_expr_dim=pert_data.adata.shape[1],        # input dimension of control_expr
    hidden_dim=128
)


optimizer = torch.optim.Adam(node_model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()

for epoch in range(10):
    node_model.train()
    total_loss = 0
    for i, (pert_emb, y) in enumerate(train_loader):
        pert_emb = pert_emb.float()
        y = y.float()
        control_expr_tensor = torch.tensor(control_expr, dtype=torch.float32)
        batch_start = i * train_loader.batch_size
        batch_end = batch_start + len(pert_emb)
        control_expr_batch = control_expr_tensor[batch_start:batch_end]
        pred = node_model(pert_emb, control_expr_batch)

        loss = loss_fn(pred, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        total_loss += loss.item()
    print(f"[Epoch {epoch+1}] Loss: {total_loss:.4f}")
```

```python
class AugmentedODEF(nn.Module):
    def __init__(self, dim, aug_dim, m_dim):
        super().__init__()
        self.dim = dim
        self.aug_dim = aug_dim
        self.total_dim = dim + aug_dim
        self.A = nn.Linear(self.total_dim, self.total_dim, bias=False)
        self.B = nn.Linear(self.total_dim, self.total_dim, bias=True)
        self.W = nn.Linear(self.total_dim, self.total_dim, bias=False)
        self.m_proj = nn.Linear(m_dim, self.total_dim)  # pert emb to augmented space

    def forward(self, t, x, m):
        m_proj = self.m_proj(m)                          # [batch, total_dim]
        activated = F.relu(self.B(x))                    # non-linear transformation
        stimulation = self.A(activated) * m_proj         # Aσ(Bx) ⊙ M
        decay = self.W(x)                                # decay term
        return stimulation - decay

class ODEWrapper(nn.Module):
    def __init__(self, odefunc, m):
        super().__init__()
        self.odefunc = odefunc
        self.m = m

    def forward(self, t, x):
        return self.odefunc(t, x, self.m)

# Perturbed Augmented Neural ODE
class PerturbedAugmentedNODE(nn.Module):
    def __init__(self, input_dim, output_dim, control_expr_dim, hidden_dim=128, aug_dim=32):
        super().__init__()
        self.control_proj = nn.Linear(control_expr_dim, input_dim)
        self.input_layer = nn.Linear(input_dim, hidden_dim)
        self.aug_dim = aug_dim
        self.odefunc = AugmentedODEF(hidden_dim, aug_dim, input_dim)
        self.output_layer = nn.Linear(hidden_dim + aug_dim, output_dim)

    def forward(self, pert_emb, control_expr):
        control_proj = self.control_proj(control_expr)
        x0 = self.input_layer(control_proj)
        aug = torch.zeros_like(x0[:, :self.aug_dim])     # augment space initialized to zero
        x0_aug = torch.cat([x0, aug], dim=1)
        ode_wrapper = ODEWrapper(self.odefunc, pert_emb)
        t = torch.tensor([0., 1.], device=x0.device)
        out = odeint(ode_wrapper, x0_aug, t, method='dopri5')[1]
        return self.output_layer(out)
    

train_dataset = PerturbationDataset(
    adata=pert_data.adata,
    set2conditions=pert_data.set2conditions,
    split='train',
    gene_emb_matrix=gene_emb_matrix,
    pert_emb_matrix=pert_emb_matrix,
    node_map=pert_data.node_map,
    node_map_pert=pert_data.node_map_pert
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

# Replace model class here
anode_model = PerturbedAugmentedNODE(
    input_dim=pert_emb_matrix.shape[1],              # perturbation embedding dim
    output_dim=pert_data.adata.shape[1],             # gene count
    control_expr_dim=pert_data.adata.shape[1],       # control expression input dim
    hidden_dim=128,
    aug_dim=32                                       # extra capacity
)

optimizer = torch.optim.Adam(anode_model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()
control_expr_tensor = torch.tensor(control_expr, dtype=torch.float32)
for epoch in range(10):
    anode_model.train()
    total_loss = 0
    for i, (pert_emb, y) in enumerate(train_loader):
        pert_emb = pert_emb.float()
        y = y.float()

        batch_start = i * train_loader.batch_size
        batch_end = batch_start + len(pert_emb)
        control_expr_batch = control_expr_tensor[batch_start:batch_end]

        pred = anode_model(pert_emb, control_expr_batch)

        loss = loss_fn(pred, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        total_loss += loss.item()
    print(f"[Epoch {epoch+1}] Loss: {total_loss:.4f}")
```

```python
# Load Augmented NODE model

class AugmentedODEF(nn.Module):
    def __init__(self, dim, aug_dim, m_dim):
        super().__init__()
        self.dim = dim
        self.aug_dim = aug_dim
        self.total_dim = dim + aug_dim
        self.A = nn.Linear(self.total_dim, self.total_dim, bias=False)
        self.B = nn.Linear(self.total_dim, self.total_dim, bias=True)
        self.W = nn.Linear(self.total_dim, self.total_dim, bias=False)
        self.m_proj = nn.Linear(m_dim, self.total_dim)  # pert emb to augmented space

    def forward(self, t, x, m):
        m_proj = self.m_proj(m)                          # [batch, total_dim]
        activated = F.relu(self.B(x))                    # non-linear transformation
        stimulation = self.A(activated) * m_proj         # Aσ(Bx) ⊙ M
        decay = self.W(x)                                # decay term
        return stimulation - decay

class ODEWrapper(nn.Module):
    def __init__(self, odefunc, m):
        super().__init__()
        self.odefunc = odefunc
        self.m = m

    def forward(self, t, x):
        return self.odefunc(t, x, self.m)

# Perturbed Augmented Neural ODE
class PerturbedAugmentedNODE(nn.Module):
    def __init__(self, input_dim, output_dim, control_expr_dim, hidden_dim=128, aug_dim=32):
        super().__init__()
        self.control_proj = nn.Linear(control_expr_dim, input_dim)
        self.input_layer = nn.Linear(input_dim, hidden_dim)
        self.aug_dim = aug_dim
        self.odefunc = AugmentedODEF(hidden_dim, aug_dim, input_dim)
        self.output_layer = nn.Linear(hidden_dim + aug_dim, output_dim)

    def forward(self, pert_emb, control_expr):
        control_proj = self.control_proj(control_expr)
        x0 = self.input_layer(control_proj)
        aug = torch.zeros_like(x0[:, :self.aug_dim])     # augment space initialized to zero
        x0_aug = torch.cat([x0, aug], dim=1)
        ode_wrapper = ODEWrapper(self.odefunc, pert_emb)
        t = torch.tensor([0., 1.], device=x0.device)
        out = odeint(ode_wrapper, x0_aug, t, method='dopri5')[1]
        return self.output_layer(out)
    

train_dataset = PerturbationDataset(
    adata=pert_data.adata,
    set2conditions=pert_data.set2conditions,
    split='train',
    gene_emb_matrix=gene_emb_matrix,
    pert_emb_matrix=pert_emb_matrix,
    node_map=pert_data.node_map,
    node_map_pert=pert_data.node_map_pert
)
anode_model = PerturbedAugmentedNODE(
    input_dim=pert_emb_matrix.shape[1],              # perturbation embedding dim
    output_dim=pert_data.adata.shape[1],             # gene count
    control_expr_dim=pert_data.adata.shape[1],       # control expression input dim
    hidden_dim=128,
    aug_dim=32                                       # extra latent dimensions
)
anode_model.load_state_dict(torch.load("anode_model.pt"))
anode_model.eval()

# Evaluate and collect predictions
anode_predictions, anode_truths = [], []

with torch.no_grad():
    for i, (pert_emb, y) in enumerate(test_loader):
        pert_emb = pert_emb.float()
        y = y.float()

        batch_start = i * test_loader.batch_size
        batch_end = batch_start + len(pert_emb)
        control_expr_batch = control_expr_tensor[batch_start:batch_end]

        pred = anode_model(pert_emb, control_expr_batch)
        anode_predictions.append(pred.cpu().numpy())
        anode_truths.append(y.cpu().numpy())

# Concatenate and compute squared error
anode_preds = np.concatenate(anode_predictions)
anode_truth = np.concatenate(anode_truths)
anode_model_errors = (anode_preds - anode_truth) ** 2  # shape: (n_samples, n_genes)
```

```python
import torch
import torch.nn as nn
import numpy as np
import pandas as pd
from torch.utils.data import DataLoader
from scipy.stats import wasserstein_distance, entropy, pearsonr

# -------- 1. Define Model --------
class LinearPertModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super().__init__()
        self.linear = nn.Linear(input_dim, output_dim)

    def forward(self, pert_emb, control_expr):
        return control_expr + self.linear(pert_emb)

# -------- 2. Training --------
train_dataset = PerturbationDataset(
    adata=pert_data.adata,
    set2conditions=pert_data.set2conditions,
    split='train',
    gene_emb_matrix=gene_emb_matrix,
    pert_emb_matrix=pert_emb_matrix,
    node_map=pert_data.node_map,
    node_map_pert=pert_data.node_map_pert
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)

model = LinearPertModel(input_dim=pert_emb_matrix.shape[1], output_dim=pert_data.adata.shape[1])
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()

prev_loss = float('inf')
threshold = 0.03
max_epochs = 30
control_expr_tensor = torch.tensor(control_expr, dtype=torch.float32)

for epoch in range(max_epochs):
    total_loss = 0
    for i, (p_emb, y) in enumerate(train_loader):
        batch_start = i * train_loader.batch_size
        batch_end = batch_start + len(p_emb)
        control_expr_batch = control_expr_tensor[batch_start:batch_end]

        p_emb = p_emb.float()
        y = y.float()
        pred = model(p_emb, control_expr_batch)

        loss = loss_fn(pred, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")
    if abs(prev_loss - total_loss) < threshold:
        print(f"Early stopping at epoch {epoch+1}")
        break
    prev_loss = total_loss

torch.save(model.state_dict(), "linear_model.pth")
print("Linear model saved to linear_model.pth")

# -------- 3. Evaluation --------
test_dataset = PerturbationDataset(
    adata=pert_data.adata,
    set2conditions=pert_data.set2conditions,
    split='test',
    gene_emb_matrix=gene_emb_matrix,
    pert_emb_matrix=pert_emb_matrix,
    node_map=pert_data.node_map,
    node_map_pert=pert_data.node_map_pert
)

test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

test_indices = test_dataset.indices if hasattr(test_dataset, "indices") else range(len(test_dataset))
control_expr_test = control_expr[test_indices]
control_expr_tensor = torch.tensor(control_expr_test, dtype=torch.float32)

model.eval()
linear_preds, linear_truth = [], []

with torch.no_grad():
    for i, (p_emb, y_true) in enumerate(test_loader):
        p_emb = p_emb.float()
        y_true = y_true.float()
        batch_start = i * test_loader.batch_size
        batch_end = batch_start + len(p_emb)
        control_expr_batch = control_expr_tensor[batch_start:batch_end]
        pred = model(p_emb, control_expr_batch)
        linear_preds.append(pred.cpu().numpy())
        linear_truth.append(y_true.cpu().numpy())

linear_preds = np.concatenate(linear_preds, axis=0)  # shape: (n_cells, n_genes)
linear_truth = np.concatenate(linear_truth, axis=0)

```
